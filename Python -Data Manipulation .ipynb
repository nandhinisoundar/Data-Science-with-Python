### 28. Get top n or bottom n rows after groupby column :

```python
def gettopn(data,col,n) :
    x = data.sort_values(by=col,ascending = False).head(n)
    return x

def getbottomn(data,col,n) :
    x = data.sort_values(by=col,ascending = True).head(n)
    return x
```

```python
Sample_SuperStore.head(5)
```

```python
Sample_SuperStore[['Region','Sales']].groupby(['Region']).apply(gettopn,col='Sales',n=3)
```

```python
Sample_SuperStore[['Region','Sales']].groupby(['Region']).apply(getbottomn,col='Sales',n=3)
```
